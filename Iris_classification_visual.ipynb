#### 1. Load the Iris dataset using seaborn or sklearn. Display the first five rows of the dataset.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
print(iris.head())
```

#### 2. Use Seaborn to plot a pairplot of the dataset, showing different species in different colors.

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

#### 3. Create a boxplot for each numerical feature (sepal length, sepal width, petal length, petal width) categorized by species.

```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='species', y=feature, data=iris)
    plt.title(f'Boxplot of {feature} by Species')
    plt.xlabel('Species')
    plt.ylabel(feature)
    plt.show()
```

#### 4. Plot a violin plot for the 'petal length' feature grouped by species.

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='species', y='petal_length', data=iris)
plt.title("Violin Plot of Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()
```

#### 5. Create a scatter plot using Matplotlib to visualize petal length vs petal width, coloring points by species.

```python
colors = {'setosa': 'blue', 'versicolor': 'orange', 'virginica': 'green'}
plt.figure(figsize=(6, 4))
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'],
                color=colors[species], label=species)
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

#### 6. Draw a histogram to show the distribution of sepal lengths for all species combined.

```python
plt.figure(figsize=(6, 4))
plt.hist(iris['sepal_length'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Sepal Lengths (All Species)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### 7. Plot a heatmap of the correlation matrix for all numerical features in the Iris dataset.

```python
plt.figure(figsize=(6, 4))
corr_matrix = iris.drop('species', axis=1).corr()
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu')
plt.title("Correlation Matrix Heatmap")
plt.show()
```
